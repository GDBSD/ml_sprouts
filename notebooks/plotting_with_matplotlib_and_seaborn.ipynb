## Univariate Plots
Some common and essential univariate plots are line plots, bar plots, histograms and density plots, and the box and whisker plot, to mention just a few.

### Line Plots
We'll use both Matplotlib and Seaborn to plot offset sine waves.

````python
def sinplot():
    """A simple plot function, to plot offset sine waves."""
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i))
````

If we call the function it will produce the default Matplotlib plot

````python
sinplot()
````

If we want to switch to seaborn defaults, we simply call Seaborn ```set``` function:

````python
sns.set()
sinplot()
````

### BAR PLOT
Let’s create a simple bar plot using the bar method.

````python
# Create some data.
states = ["Cross River", "Lagos", "Rivers", "Kano"]
population = [3737517, 17552940, 5198716, 11058300]
````

````python
# create barplot using matplotlib
plt.bar(states, population)
plt.show()
````

````python
# create barplot using seaborn
sns.barplot(x=states, y=population)
plt.show()
````

### HISTOGRAM/DENSITY PLOTS
Histogram and density plots are essential for examining the statistical distribution of a variable. For a simple histogram, we’ll create a set of 100,000 points from the normal distribution.

````python
# Create 100000 data points from the normal distributions
data = np.random.randn(100000)
````

````python
# Create a histogram plot using Matplotlib.
plt.hist(data)
plt.show()
````

````python
# Create a density plot using seaborn
my_fig = sns.distplot(data, hist=False)
plt.show()
````

### BOX AND WHISKER PLOTS
Boxplot, also popularly called box and whisker plot, is another useful visualization technique for gaining insights into the underlying data distribution. The boxplot draws a box with the upper line representing the 75th percentile and the lower line the 25th percentile. A line is drawn at the center of the box indicating the 50th percentile or median value. The whiskers at both ends give an estimation of the spread or variance of the data values. The dots at the tail end of the whiskers represent possible outlier values. 

````python
# Create data points
data = np.random.randn(1000)
````

````python
# Box plot with matplotlib
plt.boxplot(data)
plt.show()
````

````python
# Box plot with seaborn
sns.boxplot(data)
plt.show()
````

## Multivariate Plots
Common multivariate visualizations include the scatter plot and its extension the pairwise plot, parallel coordinate plots, and the covariance matrix plot.

### SCATTER PLOT
Scatter plot exposes the relationships between two variables in a dataset. 

````python
# Create the dataset
x = np.random.sample(100)
y = 0.9 * np.asarray(x) + 1 + np.random.uniform(0,0.8, size=(100,))
````

````python
# Scatter plot with matplotlib
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
# Scatter plot with seaborn
sns.regplot(x=x, y=y, fit_reg=False)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

### PAIRWISE SCATTER PLOT
Pairwise scatter plot is an effective window for visualizing the relationships among multiple variables within the same plot. However, with higher-dimension datasets, the plot may become clogged up, so use it with care. Let’s see an example of this with Matplotlib and seaborn.

Here, we will use the method scatter_matrix , one of the plotting functions in Pandas to graph a pairwise scatter plot matrix

````python
# create the dataset
data = np.random.randn(1000, 6)
````

````python
# Using Pandas scatter_matrix
pd.plotting.scatter_matrix(pd.DataFrame(data), alpha=0.5, figsize=(12, 12), diagonal="kde")
plt.show()
````

````python
# Pairwise scatter with seaborn
sns.pairplot(pd.DataFrame(data))
plt.show()
````

### CORRELATION MATRIX PLOTS
Again, correlation shows how much relationship exists between two variables. By plotting the correlation matrix, we get a visual representation of which variables in the dataset are highly correlated. Remember that parametric machine learning methods such as logistic and linear regression can take a performance hit when variables are highly correlated. Also, in practice, the correlation values that are greater than –0.7 or 0.7 are for the most part highly correlated.

````python
# Create the dataset
data = np.random.random([1000,6])
````

````python
# Plot covariance matrix using the Matplotlib matshow function
fig = plt.figure()
ax = fig.add_subplot(111)
my_plot = ax.matshow(pd.DataFrame(data).corr(), vmin=-1, vmax=1)
fig.colorbar(my_plot)
plt.show()
````

````python
# Plot covariance matrix with seaborn heatmap function
sns.heatmap(pd.DataFrame(data).corr(), vmin=-1, vmax=1)
plt.show()
````
